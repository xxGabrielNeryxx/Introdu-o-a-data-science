# Ex 9

```python
df_1 = pd.DataFrame([2, 4, 6, 8], index=['A', 'B', 'C', 'D'], columns=['pares'])
df_1
```

```python
df_2 = pd.DataFrame([1, 3, 5, 7], index=['C', 'D', 'E', 'F'], columns=['impares'])
df_2
```

```python
pd.merge(
    left = df_1,
    right = df_2, 
    left_index= True,
    right_index = True,
    how = 'inner' #Com inner apenas as colunas em comum são combinadas
)
```

```python
pd.merge(
    left = df_1,
    right = df_2, 
    left_index= True,
    right_index = True,
    how = 'outer' #Com outer todas as colunas são combinadas, e as que n forem em comum ficam  NaN
)
```

```python
pd.merge(
    left = df_1,
    right = df_2, 
    left_index= True,
    right_index = True,
    how = 'left' #Com left apenas as colunas de do parametro left=df_1 são usadas, e as que n forem em comum ficam NaN
)
```

```python
pd.merge(
    left = df_1,
    right = df_2, 
    left_index= True,
    right_index = True,
    how = 'right' #Com right apenas as colunas de do parametro right=df_2  são usadas, e as que n forem em comum ficam  NaN
)
```

# Ex 16

```python
dados = {
    'acessorios': [
        ['4 X 4', 'Câmera de estacionamento', 'Controle de tração', 'Câmbio automático'],
        ['Câmera de estacionamento', 'Câmbio Automático', 'Rodas de liga', 'Freios ABS'],
        ['Central multimídia', 'Bancos de couro', 'Rodas de liga', 'Câmera de estacionamento'],
        ['Bancos de couro', 'Freios ABS', 'Rodas de liga', 'Câmbio automático'],
        ['Central multimídia', 'ABS', 'Câmbio automático', 'Piloto automático'],
        ['Câmbio automático', 'Controle de tração', 'Piloto automático', 'Central multimídia'],
        ['4 X 4', 'Piloto automático', 'Central multimídia', 'Bancos de couro']        
    ]
}
df = pd.DataFrame(dados)
df
```

```python
acessorios = df['acessorios'].explode().str.lower()
n_acessorios = acessorios.where(
    cond = acessorios != 'abs', 
    other = 'freios abs'                      
).value_counts()
n_acessorios = n_acessorios.to_frame().rename_axis('Acessórios').rename(columns={'acessorios':'Frequencias'})
n_acessorios
```
