#### 1)Selecione somente os imóveis classificados com tipo 'Apartamento'.

```python
selecao = dados['Tipo'] == 'Apartamento'
apartamentos = dados[selecao]
apartamentos
```

```python
n1 = apartamentos.shape[0]
n1
```

#### 2)Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

```python
#NAO ESQUECER DOS PARENTESES 
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
imoveis = dados[selecao]
```

```python
n2 = imoveis.shape[0]
n2
```

#### 3)Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

```python
selecao = (dados['Area']>=60) & (dados['Area']<=100)
area = dados[selecao]
area
```

```python
n3= area.shape[0]
n3
```

#### 4)Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
selecao = (dados['Quartos']>=4) & (dados['Valor'] <2000)
quartos_4 = dados[selecao]
quartos_4
```

```python
n4 = quartos_4.shape[0]
n4
```

```python
print(f"Nº de imóveis classificados com tipo 'Apartamento' -> {n1}")
print(f"Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila' -> {n2}")
print(f"Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {n3}")
print(f"Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {n4}")
```
