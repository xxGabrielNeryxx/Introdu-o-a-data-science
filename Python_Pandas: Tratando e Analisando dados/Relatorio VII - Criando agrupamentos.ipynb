```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Media do preço do aluguel de todos imoveis 
dados['Valor'].mean()
```

```python
# #selecionando bairros especificos com isin()
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
#Criando o agrupamentos pelo Bairro
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
#Atributo .groups cria um dicionario com o nome do registro da linha associado a coluna escolhida na criaçao do grupo
#e as posiçoes em que esse elemento foi encontrado 
grupo_bairro.groups['Barra da Tijuca']
```

```python
#bairro armazena os tipos individuais de elementos da coluna 'Bairro'
for bairro, data in grupo_bairro:
    print(bairro)
```

```python
#dados armazena um dataframe para cada tipo individual de bairro

for bairro, data in grupo_bairro:
    print(type(data),data)
```

```python
#Media dos alugueis por bairro
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, round(data['Valor'].mean(),2)))
```

```python
#Media do valor dos alugueis e do condominio por bairro
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatisticas descritivas

```python
#Gera estatisticas
grupo_bairro['Valor'].describe().round(2)
```

```python
# group.aggregate(lista_com_estatisticas) permite eu escolher estatisticas
# .rename() renomeia linhas e colunas
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min':'Mínimo', 'max':'Máximo', 'sum':'Soma'}) 
```

# Criando graficos


```python
#Configuraçoes especificas para o matplotlib com juptet
%matplotlib inline 
#mostra um grafico estatico no jupter
import matplotlib.pyplot as plt
plt.rc('figure', figsize =(20,10)) #configura o tamanho de todas as figuras
```

```python
#Criando grafico do valor medio
fig = grupo_bairro['Valor'].mean().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
fig
```

```python
#Criando grafico do desvio padrao
fig = grupo_bairro['Valor'].std().plot.bar(color='blue')
fig.set_ylabel('Desvio Padrão do Aluguel')
fig.set_title('Desvio padrao do Aluguel por Bairro', {'fontsize': 22})
fig
```
